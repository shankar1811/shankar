```python
import pandas as pd
import numpy as np
data = pd.read_csv("/content/age_salary_dataset.csv")
print(data)
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
print(x)
print(y)
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy="mean")
x=imp.fit_transform(x)
y=y.reshape(-1,1)
y=imp.fit_transform(y)
y=y.reshape(-1)
print(x)
print(y)
X_df=pd.DataFrame(x,columns=[data.columns[:-1]])
Y_df=pd.DataFrame(y,columns=[data.columns[-1]])
combined_data=pd.concat([X_df,Y_df], axis=1)
print("Combined data:")
print(combined_data)
```
